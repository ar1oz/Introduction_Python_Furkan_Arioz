```python
attempt=4
def main():
    global d2
    i=0
    global attempt
    
    ask_name = input("What is your name? ")
    ask_sname=input("What is your surname? ")
    d={"name":"furkan","sname":"arioz"}
    course_list=["chem","phy","bio","math","geo"]
    course_list_user=[]
    d2={"midterm":[],"final":[],"project":[]}
    
    if ask_name==d.get('name') and ask_sname==d.get('sname'):
            print("Welcome",d.get('name'),d.get('sname'))
    else:
        attempt-=1
        print("You have entered a wrong data. You got only",+attempt,"attempt left!")
        if attempt!=0:
            main()
        else:
            print("Please try again later!")
            exit()
    while i<len(course_list)+1:
        course=input("Enter the course that you want to enroll (If you don't want to add a course, please press 'q'):")
        if course=='q':
            if i<3:
                print("You failed in class!!!")
                return 0
            elif 3<=i<=5:
                course_chosen=input("Which course's grades that you want to enter?")
                score=grade_calc()
                last_letter=grade_indic(score)
                if score<30:
                    print("You recieved FF. Therefore , you failed the course!!!")
                    exit()
                else:
                    print("You have got %s. Therefore, you have passed the %s .Congratulations!"%(last_letter,course_chosen))
                    print("Your grades are",d2)
                    return 0
            else:
                return 0
                
                
        elif course not in course_list:
            print("There is no course like this!")
            main()
        elif i!=5:
            course_list_user.append(course)
        else:
            print("You have reached the maximum courses, you cannot enter more!")
        i+=1
    exit()
def grade_calc():
    mt=float(float(input("Enter midterm score :"))*30/100)
    if 0<=mt<=100:
        final=float(float(input("Enter final score :"))*50/100)
        if 0<=final<=100:
            proj=float(float(input("Enter project score: "))*20/100)
            if 0<=proj<=100:
                d2["midterm"].append(mt)
                d2["final"].append(final)
                d2["project"].append(proj)
                return float(mt+final+proj)
            else:
                print("Please try again ")
                grade_calc()
        else:
            print("Please try again ")
            grade_calc()
    else:
        print("Please try again ")
        grade_calc()
            
 
def grade_indic(letter): 
    if letter> 90: 
        return "AA"
    elif 70<letter<90: 
        return "BB"
    elif 50<letter<70: 
        return "CC"
    elif 30<letter<50: 
        return "DD"
    elif letter<30: 
        return "FF"

    
main()
        

```
