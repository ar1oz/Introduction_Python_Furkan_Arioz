```python
def main():
    name = input("What is your name? ")
    print("Hello Mr/Ms. , " + name,)
    guess_word=input("Enter the word that is going to be found:    ")
    finding_guess = ''
    chance = 5
    while chance> 0:
        failed = 0
        for x in guess_word:
            if x in finding_guess:
                print(x)
            else:
                print("_")
                failed += 1
        if failed == 0:
            print("Won the lottery!   ")
            ans=input("Wanna play more ? (yes/no)  ")
            if ans in ['yes']:
                main()
            elif ans in ['no']:
                print("See you around, ßß!  ")
                exit()
            else:
                print("You have entered a wrong parameter!  ")
            
            break
        print('')
        info_word = input("Take a guess:   ")
        finding_guess += info_word
        if info_word not in guess_word:
            chance -= 1
            print("Wrong character   ")
            print("You have got only   ", + chance)
        if chance == 0:
            print("You Lose   ")
            ans2=input("Wanna play more ? (yes/no)  ")
            if ans2 in ['yes']:
                main()
            elif ans2 in ['no']:
                print("See you around, ßß!   ")
                exit()
            else:
                print("You have entered a wrong parameter!    ")
            
main()
```
